```python
with open("a.gdf", "r") as f:
    x = f.readlines()


nodedef=[]
ini = 0
for line in x:
    ini+=1
    if line.startswith('edgedef'):
        break
    nodedef.append(line)
    
edgedef = []
for line in x[ini-1:]:
    edgedef.append(line)
```

```python
#create dinamically using metaclasse
class Edge(object):
    def __init__(self, node1, node2, nivelOrientacao, tipoOrientacao, anoInicio, anoConclusao, node1Nome, node2Nome, directed):
        self._node1 = node1
        self._node2 = node2
        self._nivelOrientacao = nivelOrientacao
        self._tipoOrientacao = tipoOrientacao
        self._anoInicio = anoInicio
        self._anoConclusao = anoConclusao
        self._node1Nome = node1Nome
        self._node2Nome = node2Nome
        self._directed = directed
        
        @property
        def node1(self):
            return self._node1

        @node1.setter
        def node1(self, node1):
            self._node1 = node1
        
        @property
        def node2(self):
            return self._node2

        @node2.setter
        def node2(self, node2):
            self._node2 = node2
            
        @property
        def nivelOrientacao(self):
            return self._nivelOrientacao

        @nivelOrientacao.setter
        def nivelOrientacao(self, nivelOrientacao):
            self._nivelOrientacao = nivelOrientacao
            
        @property
        def tipoOrientacao(self):
            return self._tipoOrientacao

        @tipoOrientacao.setter
        def tipoOrientacao(self, tipoOrientacao):
            self._tipoOrientacao = tipoOrientacao
            
        @property
        def anoInicio(self):
            return self._anoInicio

        @anoInicio.setter
        def anoInicio(self, anoInicio):
            self._anoInicio = anoInicio
            
        @property
        def anoConclusao(self):
            return self._anoConclusao

        @anoConclusao.setter
        def anoConclusao(self, anoConclusao):
            self._anoConclusao = anoConclusao
            
        @property
        def node1Nome(self):
            return self._node1Nome

        @node1Nome.setter
        def node1Nome(self, node1Nome):
            self._node1Nome = node1Nome
        
        @property
        def node2Nome(self):
            return self._node2Nome

        @node2Nome.setter
        def node2Nome(self, node2Nome):
            self._node2Nome = node2Nome
        
        @property
        def directed(self):
            return self._directed

        @directed.setter
        def directed(self, directed):
            self._directed = directed
        


egde_list = []


for i in range(2,len(edgedef)):
    edge = edgedef[i].split(",")
    x = Edge(edge[0], edge[1], edge[2], edge[3], edge[4], edge[5], edge[6], edge[7], edge[8])
    egde_list.append(x)

p=egde_list[0]

print(p.__dict__)
print(p._node1)
```
